```python
def random_miles_generator(remaining_miles):
    
    miles = randint(40, 60)
  
    if (remaining_miles > miles)   
        return miles

    else:
        return remaining_miles
```

```python
total_distance = 1518     # distance from SPS CUNY to Key West Ferry in miles

miles_endoftheday = 0     # var, miles at the end of each day 

day_counter = 1           # var, days on a road

 
logged_miles = {}         # dict to store miles/days

while miles_endoftheday < total_distance:
    
    miles = random_miles_generator(total_distance - miles_endoftheday)
   
    logged_miles[f'Day {day_counter}'] = miles
      
    day_counter += 1
    miles_endoftheday += miles
```

```python
# pandas obj - to store miles riden each day
daily_log = pd.Series(logged_miles)

# pandas obj - to show odometer reading daily
odometer = daily_log.cumsum()
```
